```python
import json
```

```python
TRAIN_FILE = './pcqm_train_c78.json'
VAL_FILE =  './pcqm_val_c78.json'
TEST_FILE = './pcqm_test_c78.json'
```

```python
with open(TRAIN_FILE, 'rb') as f:
    train = json.load(f)
    
with open(VAL_FILE, 'rb') as f:
    val = json.load(f)
    
with open(TEST_FILE, 'rb') as f:
    test = json.load(f)
```

```python
combined_counts = {}
count = 0

for idx in range(len(train)):
    str_idx = str(idx)
    combined_counts[count] = train[str_idx]
    count += 1
```

```python
for idx in range(len(val)):
    str_idx = str(idx)
    combined_counts[count] = val[str_idx]
    count += 1
```

```python
for idx in range(len(test)):
    str_idx = str(idx)
    combined_counts[count] = test[str_idx]
    count += 1
```
